# Cosmic Microwave Background


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy
import scipy.constants as const
from scipy.optimize import curve_fit
from scipy.optimize import minimize

h = const.h
c = const.c
k = const.k

#Reading the data file

df = pd.read_table("/home/mehul/RadioKSP/Week2/COBE_CMB_data.txt",delimiter = " ")
arr = df.to_numpy()
# print(arr)

freq = []
brightness = []

for i in arr:

    freq.append(i[0])
    brightness.append(i[1])
    

freq = np.array(freq)
brightness = np.array(brightness)

# print(freq)
# print(brightness)
    
plt.scatter(freq, brightness*(10**(-3)),s = 5)
plt.ylim(0,5*(10**(-18)))

def plank(v,T):
    I = (((2*h*(v**3))/(c**2))*(1/(np.exp((h*v)/(k*T)) - 1)))
    return I

print(plank(5*(10**14),2100))

param, param_cov = curve_fit(plank, freq, brightness*(10**(-3)),2)

plt.plot(freq,plank(freq,param),color='red')

plt.xlabel('Frequency in Hz')
plt.ylabel('Brightness in SI Units')

print(param, param_cov)

plt.savefig("CMB.png")

```

# Galaxy Rotation Curve


```python
#FITTING DATASET 1
distance = []
central_line = []
sigma = []
A = []

df = pd.read_table("/home/mehul/RadioKSP/Week2/spectrum_d_0.69_kpc.txt",delimiter = " ")
arr = df.to_numpy()

wavelength = []
brightness = []

for i in arr:
#     if(i[0]>20.99 and i[0]<21.01 and i[1] > 1000):
        wavelength.append(i[0])
        brightness.append(i[1])


wavelength = np.array(wavelength)
brightness = np.array(brightness)

def Gauss(x, x0, A,sigma):
    return A*(1/(np.sqrt(2*np.pi)*sigma))*np.exp((-(x-x0)**2)/(2*sigma**2))

parameters, covariance = curve_fit(Gauss,wavelength,brightness,(21,1,1))

print(parameters)

plt.scatter(wavelength, brightness,s = 5)
plt.plot(wavelength,Gauss(wavelength,2.10024518e+01,1.04474269e+01,7.01350689e-04),color="red")

distance.append(0.69)
central_line.append(2.10024518e+01)
sigma.append(7.01350689e-04)
A.append(1.04474269e+01)
```
